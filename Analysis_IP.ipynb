```python
import scipy.io
import numpy as np
from matplotlib import pyplot as plt
import processing as pross 
```

```python

array = np.arange(1, 5.25, 0.25)  # Start at 1, end at 40.5 (to include 40), step by 0.5
Ip_power= [int(x) if x.is_integer() else x for x in array]

```

```python
DataSet_list = []
for n in Ip_power:
    try:
        DataSet_list.append(scipy.io.loadmat("IP_Complete/Ip_{}.mat".format(n)))
    except Exception as e:
        # Print the error message
        print(f"An error occurred: {e}")
        print("IP_Complete/Ip_{}.mat".format(n))
    
```

```python
print(DataSet_list[0])
```

```python
Ion_Temperature = []
Ion_Temperature_error = []
Confinement_Time = []
Confinement_Time_error = []
Density = []
Density_error = []
Tripple_product = []
Tripple_product_error =[]
Ip_power_out = []
Ip_power_out_error = []
ohmic_power_out = []
ohmic_power_out_error = []
for n in DataSet_list:
    tio = np.array(pross.get_variable(n,'tite'))*np.array(pross.get_variable(n,'te0'))
    nio = np.array(pross.get_variable(n,'ni0'))
    taue = np.array(pross.get_variable(n,"taue"))
    ip = np.array(pross.get_variable(n,"ip"))
    pohm = np.array(pross.get_variable(n,"pohm"))
    Ip_power_out.append(np.mean(ip[50:100]))
    Ip_power_out_error.append(np.mean(ip[50:100]))
    Ion_Temperature.append(np.mean(tio[50:100]))
    Ion_Temperature_error.append(np.std(tio[50:100]))
    Confinement_Time.append(np.mean(taue[50:100]))
    Confinement_Time_error.append(np.std(taue[50:100]))
    Density.append(np.mean(nio[50:100]))
    Density_error.append(np.std(nio[50:100]))
    tp = tio*nio*taue
    Tripple_product.append(np.mean(tp[50:100]))
    Tripple_product_error.append(np.std(tp[50:100]))
    ohmic_power_out.append(np.mean(pohm[30:120]))
    ohmic_power_out_error.append(np.std(pohm[30:120]))

Ion_Temperature = np.array(Ion_Temperature)
Ion_Temperature_error = np.array(Ion_Temperature_error )
Confinement_Time = np.array(Confinement_Time)
Confinement_Time_error = np.array(Confinement_Time_error)
Density = np.array(Density)
Density_error = np.array(Density_error)
Tripple_product = np.array(Tripple_product )
Tripple_product_error =np.array(Tripple_product_error)
Ip_power_out = np.array(Ip_power_out)
Ip_power_out_error = np.array(Ip_power_out_error)
ohmic_power_out = np.array(ohmic_power_out)
ohmic_power_out_error = np.array(ohmic_power_out_error)
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(Ip_power_out,Tripple_product)
ax[0,0].fill_between(Ip_power_out,Tripple_product+Tripple_product_error,Tripple_product-Tripple_product_error,alpha = 0.3)
ax[0,0].set_title("Plot of Fusion Triple Product against Current")
ax[0,0].set_ylabel("Fusion Triple Product")
ax[0,0].set_xlabel("Current (MA)")
```

```python
ax[0,1].plot(Ip_power_out,Ion_Temperature)
ax[0,1].fill_between(Ip_power_out,Ion_Temperature + Ion_Temperature_error,Ion_Temperature - Ion_Temperature_error,alpha = 0.3)
ax[0,1].set_title("Plot of Ion_Temperature against Current")
ax[0,1].set_ylabel("Ion Temperature")
ax[0,1].set_xlabel("Current (MA)")
```

```python
ax[1,0].plot(Ip_power_out,Density)
ax[1,0].fill_between(Ip_power_out,Density + Density_error,Density - Density_error,alpha = 0.3)
ax[1,0].set_title("Plot of Density against Current")
ax[1,0].set_ylabel("Density")
ax[1,0].set_xlabel("Current (MA)")
```

```python
ax[1,1].plot(Ip_power_out,Confinement_Time)
ax[1,1].fill_between(Ip_power_out,Confinement_Time+Confinement_Time_error,Confinement_Time-Confinement_Time_error,alpha = 0.3)
ax[1,1].set_title("Plot of Confinement_Time against Current")
ax[1,1].set_ylabel("Confinement_Time")
ax[1,1].set_xlabel("Current (MA)")
plt.show()
```
